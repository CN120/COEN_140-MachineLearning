```python
import numpy as np


def readFromFile(file= "spam-train"):
    y_rates = []
    feature_matrix = []
    f=open(file, "r")
    contents = f.readlines()
    for line in contents:
        y_rates.append(line[-4:-1])
        line = line[:-5]
        feature_matrix.append(line.split(','))
    return (feature_matrix, y_rates)


def normalize(matrix):
    normed = (matrix - matrix.mean(axis=0)) / matrix.std(axis=0)
    return normed.tolist()
    #returns a list
    

def Error(predictions, truths):
    total = len(truths)
    n = 0
    for i in range(total):
        if predictions[i]!=truths[i]:
            n+=1
    return float(n)/total


def Sig(n):
    return 1/(1+np.exp(np.negative(n)))
```

```python


W = None




def computeLogistic(feature_matrix, y_rates=None, training=False):
    global W,a
    if y_rates is not None:
        y_rates = np.float64(y_rates)
    feature_matrix = np.float64(feature_matrix)
    if training==True:
        W = np.ones(len(feature_matrix[0]))
        i = 0
        E = 1
        while E > 0.0001 and i < 600000:
            wt = W
            W = wt + (a * (np.dot(np.transpose(feature_matrix), (y_rates - Sig(np.dot(feature_matrix,wt))))) )
            i+=1
            E = np.linalg.norm(W-wt)
            if np.any(np.isnan(W)):
                print("W",i)
                print(wt)
                break
        print("Number of iterations: ", i)
    pred_rates = np.dot(feature_matrix, np.transpose(W))
    return pred_rates




train, y_train = readFromFile("spam-train")
train = np.float64(train)
y_train = np.float64(y_train)
train = normalize(train)


test, y_test = readFromFile("spam-test")
test = np.float64(test)
y_test = np.float64(y_test)
test = normalize(test)

total_m = train + test
for row in total_m:
    row.append(1)

x = len(train)

train = total_m[0:x]
test = total_m[x:]


a = 0.00005 #alpha
pred_train = computeLogistic(train, y_train, training = True)
pred_train_vals = []
for item in pred_train:
    if item>0:
        pred_train_vals.append(1)
    else:
        pred_train_vals.append(0)

print(Error(pred_train_vals,y_train))



pred_test = computeLogistic(test,y_test,training = False)
pred_test_vals = []
for item in pred_test:
    if item>0:
        pred_test_vals.append(1)
    else:
        pred_test_vals.append(0)

print(Error(pred_test_vals,y_test))


       
```
