```python
import numpy as np

training_data = None
testing_data = None
training_labels = None
testing_labels = None
sigma_A = np.float64(shape=(4,4))

def readFileFunction():
    f=open("iris.csv", "r")
    contents = f.readlines()
    for i in range(len(contents)):
        contents[i] = contents[i][:-1]
        contents[i] = contents[i].split(',')
        #for j in range(len(contents[i])-1):
         #   contents[i][j] = np.float64(contents[i][j])       
    return contents

def problem1():
    global training_data, testing_data, training_labels, testing_labels, sigma_A
    ret_list = readFileFunction()
    features = []
    labels = []
    for row in ret_list:
        features.append(row[:-1])
        labels.append(row[-1])
        
    training_data = np.float64(features[0:40]+features[50:90]+features[100:140])
    testing_data = np.float64(features[40:50]+features[90:100]+features[140:150])
    training_labels = np.array(labels[0:40]+labels[50:90]+labels[100:140])
    testing_labels = np.array(labels[40:50]+labels[90:100]+labels[140:150])
    

def SolveForMu(N, num_feat, class_data):
    mu = np.float64([0,0,0,0])
    for i in range(N):
        for j in range(num_feat):
            mu[j] += class_data[i][j]   #adds each row of feature data to final mu vector
    for k in range(num_feat):
        mu[k] = mu[k]/N 
    return mu


def SolveForSigma(N, num_feat, class_data, mu):
    sigma = np.zeros(shape=(num_feat, num_feat), dtype=np.float64)
    for i in range(N): # loops 40 times (aka the number of sample)
        #two loops for multiplying all values of the vector and its transpose
        for j in range(num_feat): 
            for k in range(num_feat):
                sigma[j,k]+= (class_data[i][j] - mu[j])*(class_data[i][k] - mu[k])
                # adds the value from each new matrix to sigma matrix
    for j in range(num_feat):
        for k in range(num_feat):
            sigma[j,k] = sigma[j,k]/N
    return sigma
            
# mu is a vector, sigma is a matrix, x is a vector
def LDAGausTest(mu, sigma, x):
    g = np.transpose(np.subtract(x,mu))
    g = np.dot(g,np.linalg.inv(sigma))
    g = np.dot(g,np.subtract(x,mu))
    g = -0.5*g
    return g

def QDAGausTest(mu, sigma, x):
    b = LDAGausTest(mu, sigma, x)
    a = 1.0/np.sqrt(2*np.pi*np.linalg.det(sigma))
    g = a*np.exp(b)
    return g

def LDA(data):
    mu_1 = SolveForMu(40,4,training_data[0:40])
    mu_2 = SolveForMu(40,4,training_data[40:80])
    mu_3 = SolveForMu(40,4,training_data[80:120])
    print()
    sig_1 = SolveForSigma(40,4, training_data[0:40],mu_1)
    sig_2 = SolveForSigma(40,4, training_data[40:80],mu_2)
    sig_3 = SolveForSigma(40,4, training_data[80:120],mu_3)
    sigma_A = np.add(sig_1, sig_2)
    sigma_A = np.add(sigma_A,sig_3)
    sigma_A = np.true_divide(sigma_A,3)
    
    predictions = []
    for row in data:
        p1 = LDAGausTest(mu_1,sigma_A, row)
        p2 = LDAGausTest(mu_2,sigma_A, row)
        p3 = LDAGausTest(mu_3,sigma_A, row)
        if (p1 > p2) and (p1 > p3):
            predictions.append(training_labels[0])
        elif (p2 > p1) and (p2 > p3):
            predictions.append(training_labels[41])
        else:
            predictions.append(training_labels[81])
    return predictions
            
    
    
def QDA(data):
    mu_1 = SolveForMu(40,4,training_data[0:40])
    mu_2 = SolveForMu(40,4,training_data[40:80])
    mu_3 = SolveForMu(40,4,training_data[80:120])
    print()
    sig_1 = SolveForSigma(40,4, training_data[0:40],mu_1)
    sig_2 = SolveForSigma(40,4, training_data[40:80],mu_2)
    sig_3 = SolveForSigma(40,4, training_data[80:120],mu_3)
    predictions = []
    for row in data:
        p1 = QDAGausTest(mu_1,sig_1, row)
        p2 = QDAGausTest(mu_2,sig_2, row)
        p3 = QDAGausTest(mu_3,sig_3, row)
        if (p1 > p2) and (p1 > p3):
            predictions.append(training_labels[0])
        elif (p2 > p1) and (p2 > p3):
            predictions.append(training_labels[41])
        else:
            predictions.append(training_labels[81])
    return predictions


    
```
