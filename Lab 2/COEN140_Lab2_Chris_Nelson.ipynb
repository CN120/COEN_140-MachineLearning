```python
def FillData(data, features, labels, exclude):
    for row in data:
        features.append(row[0:exclude] + row[exclude+1:-1])
        labels.append(row[-1])

def TrainAndScore(features, labels, i):
    training_data = np.array(features[0:40]+features[50:90]+features[100:140])
    training_labels = np.array(labels[0:40]+labels[50:90]+labels[100:140])
    
    testing_data = np.array(features[40:50]+features[90:100]+features[140:150])
    testing_labels = np.array(labels[40:50]+labels[90:100]+labels[140:150])
    
    lda = LinearDiscriminantAnalysis(solver="svd", store_covariance=True)
    lda.fit(training_data, training_labels)
    lda_score_train = lda.score(training_data, training_labels)
    lda_score_test = lda.score(testing_data, testing_labels)
    
    qda = QuadraticDiscriminantAnalysis(store_covariance=True)
    qda.fit(training_data, training_labels)
    qda_score_train = qda.score(training_data, training_labels)
    qda_score_test = qda.score(testing_data, testing_labels)
    
    print(f"Excluding attribute {i+1}:")
    print("-----------------------------")
    print(f"LDA training error rate = {round(1 - lda_score_train,4)}")
    print(f"LDA testing error rate = {round(1 - lda_score_test,4)}")
    print(f"QDA training error rate = {round(1 - qda_score_train,4)}")
    print(f"QDA testing error rate= {round(1 - qda_score_test,4)}")
    print()


def problem4():
    print("Baseline:")
    print("==========")
    problem2()
    problem3()
    print()
    data =  readFileFunction()
    for i in range(4):
        features = []
        labels = []
        FillData(data, features, labels, i)
        TrainAndScore(features, labels, i)
        
problem4()
```
